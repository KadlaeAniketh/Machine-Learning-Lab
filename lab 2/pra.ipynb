```python
def find_min_max(df,column_name):
    min_value=df[column_name].min()
    max_value=df[column_name].max()

    min_row=df[df[column_name]==min_value]
    max_row=df[df[column_name]==max_value]

    return min_value,max_value,min_row,max_row

hpmin,hpmax,hp_row_min,hp_row_max=find_min_max(df,"Horsepower")

print("minimum horsepower",hpmin)
print("maximum horsepower",hpmax)
print("horsepower_row_minimum_values",hp_row_min)
print("horsepower_row_miximum_values",hp_row_max)
```

```python
df['Horsepower']=pd.to_numeric(df['Horsepower'],errors='coerce')
df=df.dropna(subset=['Horsepower'])
df['Horsepower']=df['Horsepower'].astype(int)
```

```python
plt.figure(figsize=(8, 5))
plt.hist(df['Horsepower'], bins=30, density=True, alpha=0.6,
color="green", label="Histogram")
sns.kdeplot(df['Horsepower'], fill=True, color="blue", alpha=0.5,
label="KDE")
plt.xlabel("Horsepower")
plt.ylabel("Density")
plt.title(f"PDF of {'Horsepower'} (Histogram + KDE)")
plt.legend()

plt.grid()
plt.show()
```

```python
plt.figure(figsize=(8,5))
plt.hist(df['Horsepower'],bins=30,density=True,label='Histrogram',alpha=0.6,color='green')
sns.kdeplot(df['Horsepower'],fill=True,alpha=0.5,label="KDE",color="blue")

plt.xlabel('Horsepower')
plt.ylabel('Density')
plt.title(f"PDF of {'Horsepower'} (Histrogram,KDE)")

plt.grid()
plt.legend()
plt.show()
```
